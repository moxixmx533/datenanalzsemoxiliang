## Zuordnung der Aktivität in der Maschinendaten anhand der Aktivität-Daten



```python
i = 0
j = 0
while i < 3346:    ##  die len(df4)-1
    j < 196        ## die len(df2)-1
    if df4.iloc[i]['time'] < df2.iloc[j+1]['Datum']:
        i += 1
        df4.iloc[i,21] = df2.iloc[j]['Name']
    elif df4.iloc[i]['time'] >= df2.iloc[j+1]['Datum']:
        j += 1
        df4.iloc[i,21] = df2.iloc[j]['Name']
```
