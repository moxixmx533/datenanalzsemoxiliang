## Zuordnung der Aktivität in der Maschinendaten anhand der Aktivität-Daten
### Ende-Zeit in Maschinendaten ist 15:57:49. Ende-Zeit in Aktivitätdaten ist 16:50:06	.
### Start-Zeit in Maschinendaten ist 07:01:04. Ende-Zeit in Aktivitätdaten ist 07:03:08.


```python
i = 0
j = 0
while i < 32081:    ## 32081 ist die len(df)-1
    j < 216         ## 216 ist die len(df2)-1
    if df.iloc[i]['time'] < df2.iloc[j+1]['Datum']:
        i += 1
        df.iloc[i,21] = df2.iloc[j]['Name']
    elif df.iloc[i]['time'] >= df2.iloc[j+1]['Datum']:
        j += 1
        df.iloc[i,21] = df2.iloc[j]['Name']
        
```
