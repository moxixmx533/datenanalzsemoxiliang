## Zuordnung der Aktivität in der Maschinendaten anhand der Aktivität-Daten



```python
i = 0
j = 0
while i < 41152:    ##  die len(df)-1
    j < 342         ## die len(df2)-1
    if df.iloc[i]['time'] < df2.iloc[j+1]['Datum']:
        i += 1
        df.iloc[i,21] = df2.iloc[j]['Name']
    elif df.iloc[i]['time'] >= df2.iloc[j+1]['Datum']:
        j += 1
        df.iloc[i,21] = df2.iloc[j]['Name']
        
```
