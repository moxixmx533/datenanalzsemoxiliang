```python
df.drop(['Torque [%]','Boring threshold [m]','Casing Length [m]','Crowd Depth [m]','Gear Mode Main Winch []','Inclination X [Grad]','Inclination Y [Grad]','Rope Force Aux. Winch. [t]','Status Rig [ ]','Torque Steps [K=1;M=2;G=3]'], axis=1,inplace=True)
```

```python
Class = df.iloc[:,11].unique()
Class
```

```python
Class_dict = dict(zip(Class, range(len(Class))))
Class_dict
```

```python
df['Aktivität mit Nr'] = df.iloc[:,11].apply(lambda x: Class_dict[x])
df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
```

```python
lb = LabelBinarizer()
lb.fit(list(Class_dict.values()))
transformed_labels = lb.transform(df['Aktivität mit Nr'])
y_bin_labels = []  # 对多分类进行0-1编码的变量
for i in range(transformed_labels.shape[1]):
    y_bin_labels.append('y' + str(i))
    df['y' + str(i)] = transformed_labels[:, i]
```

```python
df[y_bin_labels]
```

```python
X = df.iloc[:,0:10]
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, df[y_bin_labels], test_size = 0.2, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer 
classifier.add(Dense(units = 7, activation = 'relu', kernel_initializer = 'uniform', input_dim = 10))
# units 表示这层输出的神经元个数 (10+5)/2= 8  通常设置为输入层的神经元个数加上输出层的神经元个数的平均值
# Adding the second hidden layer
classifier.add(Dense(units = 7, activation = 'relu', kernel_initializer = 'uniform'))

# Adding the output layer  这里用Oneencoder 了看来
classifier.add(Dense(units = 5, activation = 'softmax', kernel_initializer = 'uniform')) #softmax 是输出类别多个的活动函数

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy']) 
# crossentropy 是损失函数的一种

# Fitting the ANN to the Training set    batch_size 表示一组数据里包含多少个观察数据 epochs 表示这个神经网络要经行多少期的训练 
# 这两个参数都调参都比较艺术，玄学
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

### Evaluation on test data

```python
#4.Evaluation on Test data
eval = classifier.evaluate(X_test, y_test, verbose=0)
print("Evaluation on test data: loss = %0.6f accuracy = %0.2f%% \n" \
      % (eval[0], eval[1] * 100) )
```

```python
y_pred = classifier.predict(X_test)
```

```python
for i in range(len(y_pred)):
    max_value=max(y_pred[i])
    for j in range(len(y_pred[i])):
        if max_value==y_pred[i][j]:
            y_pred[i][j]=1
        else:
            y_pred[i][j]=0
```
