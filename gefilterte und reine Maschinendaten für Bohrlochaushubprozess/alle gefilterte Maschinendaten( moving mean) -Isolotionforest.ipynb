```python
df_leeren = df.loc[df['activity from avtivitydaten']=='Nebenprozesszeit - Leeren']
df_leeren.reset_index(drop=True, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

```python
from sklearn.ensemble import IsolationForest
```

```python
data = df_leeren[['Crowd Depth [m]', 'Crowd-Force [t]', 'Depth [m]','Gear Mode Main Winch []','Inclination X [Grad]','Inclination Y [Grad]','Main Winch Rope Speed [cm/min]','Pressure Pump 1 [bar]','Pressure Pump 2 [bar]','Pressure Pump 3 [bar]','Pressure Pump 4 [bar]','Rope Force Aux. Winch. [t]','Rope Force Main Winch [t]','Speed of Rotary Drive [U/min]','Torque Steps [K=1;M=2;G=3]','Torque [kNm]']]
scaler = StandardScaler()
np_scaled = scaler.fit_transform(data)
data = pd.DataFrame(np_scaled)
# 训练孤立森林模型
model =  IsolationForest(contamination=0.1)
model.fit(data)
  
#返回1表示正常值，-1表示异常值
df_leeren['anomaly2'] = pd.Series(model.predict(data))
  
fig, ax = plt.subplots(figsize=(20,15))
a = df_leeren.loc[df_leeren['anomaly2'] == -1, ['Crowd Depth [m]', 'Crowd-Force [t]', 'Depth [m]','Gear Mode Main Winch []','Inclination X [Grad]','Inclination Y [Grad]','Main Winch Rope Speed [cm/min]','Pressure Pump 1 [bar]','Pressure Pump 2 [bar]','Pressure Pump 3 [bar]','Pressure Pump 4 [bar]','Rope Force Aux. Winch. [t]','Rope Force Main Winch [t]','Speed of Rotary Drive [U/min]','Torque Steps [K=1;M=2;G=3]','Torque [kNm]']] #异常值
ax.plot(df_leeren.index, df_leeren['Depth [m]'], color='blue', label = 'normal')
ax.scatter(a.index,a['Depth [m]'], color='red', label = 'outlier')
plt.legend()
plt.show();
```

```python
fig, ax = plt.subplots(figsize=(20,15))
ax.plot(df_leeren.index, df_leeren['Torque [kNm]'], color='blue', label = 'normal')
ax.scatter(a.index,a['Torque [kNm]'], color='red', label = 'outlier')
```
